```python
from sklearn import preprocessing


label_encoder = preprocessing.LabelEncoder()


df['Gender']= label_encoder.fit_transform(df['Gender'])

df['Gender'].unique()
```

```python
label_encoder = preprocessing.LabelEncoder()


df['Location']= label_encoder.fit_transform(df['Location'])

df['Location'].unique()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Churn',axis=1),
                                                    df['Churn'], test_size=0.05, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
#results for normal logistic regression
```

```python
from sklearn.tree import DecisionTreeClassifier

clf_entropy = DecisionTreeClassifier(
            criterion = "entropy", random_state = 100,
            max_depth = 20, min_samples_leaf = 10)

    # Performing training
clf_entropy.fit(X_train, y_train)
y_pred = clf_entropy.predict(X_test)
```

```python
print(classification_report(y_pred,predictions))
#decision tree classifier results
```

```python
from xgboost import XGBClassifier
model = XGBClassifier()
model.fit(X_train, y_train)
```

```python
y_predxg = model.predict(X_test)
predictions = [round(value) for value in y_predxg]
print(classification_report(y_predxg,predictions))
```
