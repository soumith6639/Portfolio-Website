```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download the VADER lexicon a collection of words and their associated sentiment scores
nltk.download('vader_lexicon')

# Create a SentimentIntensityAnalyzer object
sia = SentimentIntensityAnalyzer()

# Sample text
text = "I love this ! It's amazing."
# text = "I hate this ! It's terrible."

# Analyze sentiment
sentiment_scores = sia.polarity_scores(text)

# Interpret sentiment scores
compound_score = sentiment_scores['compound']

if compound_score >= 0.05:
    sentiment = "Positive"
elif compound_score <= -0.05:
    sentiment = "Negative"
else:
    sentiment = "Neutral"

# Print results
print("Text:", text)
print("Sentiment:", sentiment)
print("Compound Score:", compound_score) # A compound score that represents the overall sentiment of the text



```

```python
from textblob import TextBlob

# Sample text
# text = "I love this ! It's fantastic."
text = 'The weather today is quite gloomy and disappointing.'

# Create a TextBlob object
blob = TextBlob(text)

# Perform sentiment analysis
sentiment = blob.sentiment

# Get sentiment polarity (-1 for negative, 1 for positive)
polarity = sentiment.polarity

# Interpret sentiment
if polarity > 0:
    sentiment_label = "Positive"
elif polarity < 0:
    sentiment_label = "Negative"
else:
    sentiment_label = "Neutral"

# Print results
print("Text:", text)
print("Sentiment Label:", sentiment_label)
print("Sentiment Polarity:", polarity)

```

```python
import nltk
from nltk.corpus import wordnet
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
# Sample paragraph
paragraph = "The generated recommendations are printed to the console."

# Tokenize the paragraph into sentences
sentences = sent_tokenize(paragraph)

# Initialize the WordNet lemmatizer
lemmatizer = WordNetLemmatizer()

# Function to get synonyms of a word
def get_synonyms(word):
    synonyms = set()
    for syn in wordnet.synsets(word):
        for lemma in syn.lemmas():
            synonyms.add(lemma.name())
    return list(synonyms)

# Function to paraphrase a sentence
def paraphrase_sentence(sentence):
    tokens = word_tokenize(sentence)
    paraphrased_tokens = []

    for token in tokens:
        # Lemmatize the token
        lemma = lemmatizer.lemmatize(token, pos='v')  # Use verb lemma
        synonyms = get_synonyms(lemma)

        # Choose a synonym if available, or keep the original token
        paraphrased_token = random.choice(synonyms) if synonyms else token
        paraphrased_tokens.append(paraphrased_token)

    # Reconstruct the paraphrased sentence
    paraphrased_sentence = ' '.join(paraphrased_tokens)

    return paraphrased_sentence

import random

# Paraphrase each sentence in the paragraph
paraphrased_paragraph = [paraphrase_sentence(sentence) for sentence in sentences]

# Join the paraphrased sentences into a new paragraph
paraphrased_paragraph = ' '.join(paraphrased_paragraph)

# Print the original and paraphrased paragraphs
print("Original Paragraph:")
print(paragraph)
print("\nParaphrased Paragraph:")
print(paraphrased_paragraph)
```

```python
!pip install transformers
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load the pre-trained GPT-2 model and tokenizer
model_name = "gpt2"  # You can choose other variants like "gpt2-medium," "gpt2-large," etc.
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

# Set the model to evaluation mode
model.eval()

# Define a function to chat with the chatbot
def chat_with_bot(prompt):
    input_ids = tokenizer.encode(prompt, return_tensors="pt")

    # Generate a response from the model
    with torch.no_grad():
        response = model.generate(input_ids, max_length=100, num_return_sequences=1, no_repeat_ngram_size=2)

    bot_response = tokenizer.decode(response[0], skip_special_tokens=True)
    return bot_response

# Main chat loop
print("ChatGPT: Hello! How can I assist you today?")
while True:
    user_input = input("You: ")

    if user_input.lower() == 'exit':
        print("ChatGPT: Goodbye!")
        break

    bot_response = chat_with_bot("You: " + user_input + "\nChatGPT:")
    print("ChatGPT:", bot_response)
```
